```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
import matplotlib.pyplot as plt

from scipy.spatial.distance import cdist
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from scipy.stats import pearsonr, spearmanr
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import silhouette_samples, silhouette_score, mean_squared_error




plt.rcParams['image.cmap'] = "bwr"
plt.rcParams['savefig.bbox'] = "tight"
plt.style.use('ggplot')
plt.rcParams['image.cmap'] = "bwr"
plt.rcParams['savefig.bbox'] = "tight"
plt.style.use('ggplot')
```

```python
covid.head()
```

```python
correlacion_covid = covid.corr()
correlacion_covid
```

```python
sns.heatmap(correlacion_covid.abs())
```

```python
sns.regplot(x="Recovered", y="Active", data=correlacion_covid)
```

```python
corr, pvalue = pearsonr(
  x = covid['Recovered'], 
  y = covid['Active'])
print("Coeficiente de correlacin de Pearson:", corr) 
print("P-value:", pvalue) 
```

```python
X = covid[['Recovered']]
y = covid['Active']

X_train, X_test, y_train, y_test = train_test_split(
  X.values.reshape(-1,1),
  y.values,
  train_size = 0.5,
  random_state = 42,  
  shuffle = True
)
```

```python
modelo = LinearRegression()
modelo.fit(X = X_train, y = y_train)
print("Ordenada:", modelo.intercept_)
print("Pendiente:", list(zip(X.columns, modelo.coef_.flatten())))
```

```python
print("Coeficiente de determinacin R:", modelo.score(X_test, y_test))
```

```python
y_pred = modelo.predict(X = X_test)

rmse = mean_squared_error(
  y_true  = y_test,
  y_pred  = y_pred,
  squared = False
)

print("RMSE:", rmse)
```

```python
def normalizar_rmse(rmse, y):
  return rmse / (max(y) - min(y))
```

```python
normalizar_rmse(rmse, covid['Active'])
```
